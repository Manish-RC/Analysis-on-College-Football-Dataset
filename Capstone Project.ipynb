```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("D:\College-Football-Original-Dataset.csv")
df.head()
```

```python
df.columns
```

```python
df.columns = ['UniversityName','Year','TeamID','CMP','ATT','YDS','CMPPercent','YPA','LNG','TD',
              'INT','SACK','SYL','RTG','R_ATT','R_AVG','TOTAL_PLAYS','RUN_PERCENT','PASS_PERCENT']
df.columns
```

```python
df.head()
```

```python
"""We can identifty that TeamID is No of much meaningful feature to consider as it it a 
   combined version of UniversityName and Year.
   Hence, it can be removed.......   
"""
df.drop(["TeamID"], axis=1,inplace=True)
```

```python
df=df.drop(0)
df.head()
```

```python
df.head
```

```python
df['UniversityName'].unique()
```

```python
df.describe()
```

```python
df.info()
```

```python
'''
Noticing that CMPPercent,YPA,RTG,R_AVG,RUN_PERCENT,PASS_PERCENT are float values.
Hence needed to be converted from object into float.
'''
df['CMPPercent'] = pd.to_numeric(df['CMPPercent'], errors='coerce')
df['YPA'] = pd.to_numeric(df['YPA'], errors='coerce')
df['RTG'] = pd.to_numeric(df['RTG'], errors='coerce')
df['R_AVG'] = pd.to_numeric(df['R_AVG'], errors='coerce')
```

```python
'''
Before onverting string into integer notice column "YDS" has comma in between the numbers.
Hence need to be removed before conversion into integer.
'''
df["YDS"] = df["YDS"].str.replace(",","")
df['YDS'] = df['YDS'].astype(int)
df.info()
```

```python
'''
Similarly 'RUN_PERCENT' & 'PASS_PERCENT' have '%' in them which is to be removed.
'''
df["RUN_PERCENT"] = df["RUN_PERCENT"].str.replace("%","")
df["PASS_PERCENT"] = df["PASS_PERCENT"].str.replace("%","")
df['RUN_PERCENT'] = pd.to_numeric(df['RUN_PERCENT'], errors='coerce')
df['PASS_PERCENT'] = pd.to_numeric(df['PASS_PERCENT'], errors='coerce')
```

```python
#now converting integer objects into integers
'CMP', 'ATT', 'YDS', 'LNG', 'TD', 'INT', 'SACK', 'SYL', 'R_ATT','TOTAL_PLAYS'
df['Year'] = df['Year'].astype(int)
df['CMP'] = df['CMP'].astype(int)
df['ATT'] = df['ATT'].astype(int)
df['YDS'] = df['YDS'].astype(int)
df['LNG'] = df['LNG'].astype(int)
df['TD'] = df['TD'].astype(int)
df['INT'] = df['INT'].astype(int)
df['SACK'] = df['SACK'].astype(int)
df['SYL'] = df['SYL'].astype(int)
df['R_ATT'] = df['R_ATT'].astype(int)
df['TOTAL_PLAYS'] = df['TOTAL_PLAYS'].astype(int)
df.info()
```

```python
fig, ax= plt.subplots()
ax.boxplot(df['RUN_PERCENT'], showmeans=True)

stats = {'median': np.median(df['RUN_PERCENT']),
         'q1': np.percentile(df['RUN_PERCENT'], 25),
         'q3': np.percentile(df['RUN_PERCENT'], 75),
         'min': np.min(df['RUN_PERCENT']),
         'max': np.max(df['RUN_PERCENT'])}

for i, (key, val) in enumerate(stats.items()):
    ax.text(1.1, 0.9-i*0.1, f"{key.capitalize()}: {val:.2f}", 
            transform=ax.transAxes, ha='left', va='top')
plt.show()
```

```python
#fitting linear regression
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

X = data[['CMP', 'YDS', 'YPA','R_ATT', 'R_AVG']].values
tX = data[['CMP', 'YDS', 'YPA','R_ATT', 'R_AVG']].values
# extract the dependent variable
y = data['RUN_PERCENT'].values
ty = data['RUN_PERCENT'].values

# create the linear regression model
model = LinearRegression()

# fit the model to the data
model.fit(X, y)

# print the coefficients
print('Coefficients:', model.coef_)

y_p = model.predict(tX)

# calculate the mean squared error of the model on the testing data
mse = mean_squared_error(ty, y_p)
print('Mean Squared Error:', mse)
```

```python
#applying Random Forest on data
from sklearn.ensemble import RandomForestRegressor
# Create a Random Forest model with default parameters
model = RandomForestRegressor()

# Fit the model to the training data
model.fit(X, y)

# Make predictions on the test data
y_pred = model.predict(tX)

# Calculate the Mean Squared Error
mse = mean_squared_error(ty, y_pred)
print("Mean Squared Error:", mse)
```
